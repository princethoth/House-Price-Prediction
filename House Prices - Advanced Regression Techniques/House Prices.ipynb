```python
fill_none = [
    'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
    'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'BsmtQual', 'BsmtCond'
]
for col in fill_none:
    df[col] = df[col].fillna('None')


fill_zero = ['GarageYrBlt']
for col in fill_zero:
    df[col] = df[col].fillna(0)


df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())
df['MasVnrArea'] = df['MasVnrArea'].fillna(df['MasVnrArea'].median())

df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])
```

```python
X = df.drop('SalePrice', axis=1)
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
cat_cols = X_train.select_dtypes(include=['object']).columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
```

```python
num_cols = X_train.select_dtypes(include=['int64', 'float64']).columns
cat_cols = X_train.select_dtypes(include=['object']).columns
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', MinMaxScaler(), num_cols),                  # scale numeric features
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)  # encode categorical features
    ]
)
```

```python
X_train = preprocessor.fit_transform(X_train)
```

```python
X_test = preprocessor.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential ()

model.add(Dense(277, activation = 'relu'))
model.add(Dense(277, activation = 'relu'))
model.add(Dense(277, activation = 'relu'))
model.add(Dense(277, activation = 'relu'))
model.add(Dense(1))

model.compile(optimizer = 'adam', loss = 'mse')
```

```python
model.fit(x = X_train, y = y_train, validation_data =(X_test, y_test),
          batch_size = 128, epochs = 400)
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
plt.show(1)
```

```python
from tensorflow.keras.models import load_model
import joblib

joblib.dump(preprocessor, 'preprocessor.pkl')
model.save('myhouseprice.h5')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_absolute_error(y_test, predictions)
```

```python
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
explained_variance_score(y_test, predictions)
```
